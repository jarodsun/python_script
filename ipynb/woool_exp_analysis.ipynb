```python
import matplotlib.pyplot as plt

# 等级和对应的经验值
levels = list(range(1, 51))
experience = [
    100, 200, 300, 400, 600, 900, 1200, 1700, 2500, 6000,
    8000, 10000, 15000, 30000, 40000, 50000, 70000, 100000,
    120000, 140000, 250000, 300000, 350000, 400000, 500000,
    700000, 1000000, 1400000, 1800000, 2000000, 2400000,
    2800000, 3200000, 3600000, 4000000, 4800000, 5600000,
    8200000, 9000000, 12000000, 16000000, 30000000, 50000000,
    80000000, 120000000, 480000000, 1000000000, 3000000000,
    3500000000, 4000000000
]

# 第一张图：Level1到Level10
plt.figure(figsize=(10, 5))
plt.scatter(levels[:10], experience[:10], color='blue')
plt.plot(levels[:10], experience[:10], color='red', linewidth=2)
plt.title('Experience Curve (Level 1-10)')
plt.xlabel('Level')
plt.ylabel('Experience')
plt.grid(True)
plt.show()

# 第二张图：Level11到Level30
plt.figure(figsize=(10, 5))
plt.scatter(levels[10:30], experience[10:30], color='blue')
plt.plot(levels[10:30], experience[10:30], color='red', linewidth=2)
plt.title('Experience Curve (Level 11-30)')
plt.xlabel('Level')
plt.ylabel('Experience')
plt.grid(True)
plt.show()

# 第三张图：Level31到Level50
plt.figure(figsize=(10, 5))
plt.scatter(levels[30:50], experience[30:50], color='blue')
plt.plot(levels[30:50], experience[30:50], color='red', linewidth=2)
plt.title('Experience Curve (Level 31-50)')
plt.xlabel('Level')
plt.ylabel('Experience')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# 给定的经验值
experience_level_45 = 120000000  # Level 45 的经验值
experience_level_60 = 4000000000  # Level 60 的经验值

# 等级差
level_difference = 60 - 45

# 计算增长因子 k
k = (experience_level_60 / experience_level_45) ** (1 / level_difference)

# 打印增长因子 k
print(f"The growth factor k is: {k:.4f}")

# 计算45级到60级的经验值
levels = list(range(45, 61))
experience_values = [experience_level_45]
current_experience = experience_level_45

print("Experience values from Level 45 to Level 60:")
for level in levels[1:]:  # 从第46级开始计算
    current_experience = current_experience * k  # 计算下一级的经验值
    experience_values.append(current_experience)
    # 保留数值的前2位，第三位向上取整
    display_experience = math.ceil(current_experience / 10**(len(str(int(current_experience))) - 2)) * 10**(len(str(int(current_experience))) - 2)
    print(f"Level {level}: {display_experience}")

# 绘制折线图
plt.figure(figsize=(10, 5))
plt.plot(levels, experience_values, marker='o', linestyle='-', color='b')
plt.title('Experience Curve from Level 45 to Level 60')
plt.xlabel('Level')
plt.ylabel('Experience')
plt.grid(True)
plt.show()
```
